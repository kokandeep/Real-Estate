```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn import datasets, linear_model, metrics
from sklearn.metrics import r2_score
import joblib
```

```python
df = pd.read_csv('Real estate.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['X1 transaction date'],inplace=True,axis=1)
```

```python
df
```

```python
df.drop('No',inplace=True,axis=1)
```

```python
df.rename(columns = {'X2 house age':'house_age','X3 distance to the nearest MRT station':'market_distance','X4 number of convenience stores':'no_of_conveniance_store','X5 latitude':'lat','X6 longitude':'long'}, inplace = True)
df
```

```python
y=df.pop('Y house price of unit area')
X=df
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
reg = linear_model.LinearRegression()
```

```python
lr=reg.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
y_pred=lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
joblib.dump(lr, 'estate.pkl')
```
